```python
import ee
import geemap as gee

# Authenticate and initialize Earth Engine
ee.Authenticate()
ee.Initialize()
```

```python

# Define your shapefile asset ID
shapefile_asset_id = 'projects/ee-menwagaw/assets/shp1'

# Load the shapefile as a FeatureCollection
shapefile = ee.FeatureCollection(shapefile_asset_id)

# Define the ROI using the shapefile geometry
roi = shapefile.geometry()

# Load the ERA5 LAND monthly aggregate image collection and select the temperature variable
era = ee.ImageCollection("ECMWF/ERA5_LAND/MONTHLY_AGGR").select('temperature_2m').filterDate('1980','2024')

# Convert temperature from Kelvin to Celsius
temp = era.map(lambda img: img.subtract(273.15).copyProperties(img, ['system:time_start']))

# Function to extract temperature values and dates
def ext_val(img):
    temp_val = img.reduceRegion(reducer=ee.Reducer.first(), geometry=roi, scale=10000).values().get(0)
    date_val = img.date().format('YYYY-MM-dd')
    return ee.Feature(None, {'date': date_val, 'temp': temp_val})

# Map the function over the image collection
temp_date_val = temp.map(ext_val)

# Convert the resulting FeatureCollection to a list
data_list = temp_date_val.toList(temp_date_val.size()).getInfo()

# Extract temperature and date values
temp_list = [item['properties']['temp'] for item in data_list]
date_list = [item['properties']['date'] for item in data_list]

# Create a Pandas DataFrame
import pandas as pd
df = pd.DataFrame({'date': date_list, 'temp': temp_list}).round(2)

# Save the DataFrame to a CSV file
df.to_csv('temp_dataset.csv')

# Convert the date column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Resample the data annually and calculate the mean
df_annual = df.resample('YE', on='date').mean().round(2)

# Plot the annual mean temperature
from matplotlib import pyplot as plt
df_annual['temp'].plot(kind='line', figsize=(8, 4), title='Annual Mean Temperature')
plt.gca().spines[['top', 'right']].set_visible(False)
plt.show()

# Install pymannkendall if not already installed
!pip install pymannkendall

# Perform the Mann-Kendall test
import pymannkendall as mk
mk_test = mk.original_test(df_annual['temp'])
print(mk_test)

# Install pyhomogeneity if not already installed
!pip install pyhomogeneity

# Perform the Pettitt test
import pyhomogeneity as ph
pet_test = ph.pettitt_test(df_annual['temp'])
print(pet_test)
```

```python
# Spatial Map of Annual average Temperature in Lombardy Region
import ee
import geemap
import pandas as pd
import matplotlib.pyplot as plt

# Define your shapefile asset ID
shapefile_asset_id = 'projects/ee-menwagaw/assets/shp1'

# Load the shapefile as a FeatureCollection
shapefile = ee.FeatureCollection(shapefile_asset_id)

# Define the ROI using the shapefile geometry
roi = shapefile.geometry()

# Load the ERA5 LAND monthly aggregate image collection and select the temperature variable
era = ee.ImageCollection("ECMWF/ERA5_LAND/MONTHLY_AGGR").select('temperature_2m').filterDate('1980', '2024')

# Convert temperature from Kelvin to Celsius
temp = era.map(lambda img: img.subtract(273.15).copyProperties(img, ['system:time_start']))

# Calculate the annual mean temperature
annual_temp = temp.mean()

# Clip the annual mean temperature to the ROI
annual_temp_clipped = annual_temp.clip(roi)

# Create a map object
Map = geemap.Map()

# Add the clipped temperature layer to the map
vis_params = {
    'min': -30,
    'max': 30,
    'palette': ['blue', 'cyan', 'green', 'yellow', 'red']
}
Map.addLayer(annual_temp_clipped, vis_params, 'Annual Mean Temperature')

# Center the map on the ROI
Map.centerObject(roi, 10)

# Display the map
Map
```

```python
# Trend Analysis
import ee
import geemap
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Define your shapefile asset ID
shapefile_asset_id = 'projects/ee-menwagaw/assets/shp1'

# Load the shapefile as a FeatureCollection
shapefile = ee.FeatureCollection(shapefile_asset_id)

# Define the ROI using the shapefile geometry
roi = shapefile.geometry()

# Load the ERA5 LAND monthly aggregate image collection and select the temperature variable
era = ee.ImageCollection("ECMWF/ERA5_LAND/MONTHLY_AGGR").select('temperature_2m').filterDate('1980', '2024')

# Convert temperature from Kelvin to Celsius
temp = era.map(lambda img: img.subtract(273.15).copyProperties(img, ['system:time_start']))

# Calculate the annual mean temperature
annual_temp = temp.mean()

# Clip the annual mean temperature to the ROI
annual_temp_clipped = annual_temp.clip(roi)

# Create a map object
Map = geemap.Map()

# Add the clipped temperature layer to the map
vis_params = {
    'min': -30,
    'max': 30,
    'palette': ['blue', 'cyan', 'green', 'yellow', 'red']
}
Map.addLayer(annual_temp_clipped, vis_params, 'Annual Mean Temperature')

# Center the map on the ROI
Map.centerObject(roi, 10)

# Display the map
Map

# Function to extract temperature values and dates
def ext_val(img):
    temp_val = img.reduceRegion(reducer=ee.Reducer.first(), geometry=roi, scale=10000).values().get(0)
    date_val = img.date().format('YYYY-MM-dd')
    return ee.Feature(None, {'date': date_val, 'temp': temp_val})

# Map the function over the image collection
temp_date_val = temp.map(ext_val)

# Convert the resulting FeatureCollection to a list
data_list = temp_date_val.toList(temp_date_val.size()).getInfo()

# Extract temperature and date values
temp_list = [item['properties']['temp'] for item in data_list]
date_list = [item['properties']['date'] for item in data_list]

# Create a Pandas DataFrame
df = pd.DataFrame({'date': date_list, 'temp': temp_list}).round(2)

# Save the DataFrame to a CSV file
df.to_csv('temp_dataset.csv')

# Convert the date column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Resample the data annually and calculate the mean
df_annual = df.resample('YE', on='date').mean().round(2)

# Perform trend analysis (linear regression)
X = np.array(df_annual.index.year).reshape(-1, 1)
y = df_annual['temp'].values
model = LinearRegression().fit(X, y)
trend_line = model.predict(X)

# Calculate Sen's slope
slope = model.coef_[0]
intercept = model.intercept_

# Plot the annual mean temperature with the trend line
plt.figure(figsize=(10, 6))
plt.plot(df_annual.index, df_annual['temp'], label='Annual Mean Temperature')
plt.plot(df_annual.index, trend_line, label=f'Trend Line (slope={slope:.2f}°C/year)', color='red')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')
plt.title('Annual Mean Temperature and Trend Line')
plt.legend()
plt.grid(True)
plt.show()

# Perform the Mann-Kendall test
import pymannkendall as mk
mk_test = mk.original_test(df_annual['temp'])
print(mk_test)

# Perform the Pettitt test
import pyhomogeneity as ph
pet_test = ph.pettitt_test(df_annual['temp'])
print(pet_test)
```
